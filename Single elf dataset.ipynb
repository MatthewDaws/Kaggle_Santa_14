Implement the following simple algorithm:
   - Have a "target" recharge at which we'll attempt the longest toy left.
   - Then work on tasks, as close as possible to using up the whole day, to "recharge" the elf.

```python
import math

# Time utilities
def move_to_next_sanctioned_hour(time):
    """Return `time` if in sanctioned time, or moves to 9am next day if not."""
    day_time = time % elf.minsday
    if day_time >= elf.dayends:
        return time + elf.daystarts + elf.minsday - day_time
    if day_time < elf.daystarts:
        return time + elf.daystarts - day_time
    return time

def pretty_print_time(time):
    """Print as days::hours::minutes"""
    mins = time % 60
    hours = ((time - mins)//60) % 24
    days = (time - mins - hours*60) // elf.minsday
    return "{}:{}:{}".format(days,hours,mins)

# Logging class
import collections
LogEntry = collections.namedtuple("LogEntry", ["time", "duration", "productivity"])
```

```python
# Simple class to wrap around a list
class TaskList:
    def __init__(self, alltasks):
        self.tasks = alltasks[:] # Take a copy
        self.tasks.sort()
    def tasks_remaining(self):
        return len(self.tasks)
    def longest(self):
        return self.tasks[-1]
    def shortest(self):
        return self.tasks[0]
    def remove_task(self, length):
        self.tasks.remove(length)
    def find_closest_leq(self, target):
        """Return the length of the task which is closest but <= target."""
        if self.tasks[0] > target: return None
        if self.tasks[-1] <= target: return self.tasks[-1]
        # Binary search
        index_low = 0 # Maintain invariant that self.tasks[index_low] <= target
        index_high = len(self.tasks) - 1 # and self.tasks[index_high] > target
        while index_high - index_low >= 2:
            index_mid = (index_low + index_high) // 2
            if self.tasks[index_mid] <= target:
                index_low = index_mid
            else:
                index_high = index_mid
        return self.tasks[index_low]
    def find_closest_geq(self, target):
        """Return the length of the task which is closest but >= target."""
        if self.tasks[-1] < target: return None
        if self.tasks[0] >= target: return self.tasks[0]
        index_low, index_high = 0, len(self.tasks) - 1   # tasks[index_low] < target
        while index_high - index_low >= 2:               # tasks[index_high] >= target
            index_mid = (index_low + index_high) // 2
            if self.tasks[index_mid] < target:
                index_low = index_mid
            else:
                index_high = index_mid
        return self.tasks[index_high]
```

```python
def algo_one(e, toys, log, target):
    while toys.tasks_remaining() > 0:
        if e.productivity < target(toys.longest()): # Recharge
            # If outside of sanctioned hours push to next day
            e.available_to_work = move_to_next_sanctioned_hour(e.available_to_work)
            # Time left in current day to work
            time = elf.dayends - ( e.available_to_work % elf.minsday )
            # Adjust to toy duration elf could complete in this time
            # TODO: Could allow some "wiggle room" here to overflow slightly,
            #   as that won't cost much in productivity, but might allow better
            #   allocation of tasks.
            time = int(math.floor(time*e.productivity))
            if time < 1 or time < toys.shortest():
                # Push to next day
                e.available_to_work = ( elf.daystarts +
                    elf.minsday * (1 + e.available_to_work // elf.minsday) )
                time = elf.dayends - elf.daystarts
                time = int(math.floor(time*e.productivity))
            if time < toys.shortest():
                # Still no tasks.  Just do longest remaining
                #print("Ran out of 'small' tasks with prod =",e.productivity)
                task = toys.longest()
            else:
                task = toys.find_closest_leq(time)
                if task == None:
                    raise Exception("Arse!")
        else:
            task = toys.longest() # Just do longest task
        log.append( LogEntry(e.available_to_work, task, e.productivity) )
        e.do_task(e.available_to_work,task)
        toys.remove_task(task)

def target(size):
    #return 0.4 # 410 years
    #return max(0.25, 0.5 * size / 22300) # 398!
    return 0.25 * max(1.0, size / 11500) # 397
    #return 0.25 * max(1.0, size / 12600) For variant

e = elf.Elf(1)
toys = TaskList(toys_list)
log = []
algo_one(e, toys, log, target)

print("Mins:",e.available_to_work, "Millions of Mins:",e.available_to_work // 1000000,
      "Years {:.2f}:".format(e.available_to_work / elf.minsday / 365))
```

```python
def algo_two(e, toys, log, target):
    while toys.tasks_remaining() > 0:
        if e.productivity < target(toys.longest()): # Recharge
            # If outside of sanctioned hours push to next day
            e.available_to_work = move_to_next_sanctioned_hour(e.available_to_work)
            # Time left in current day to work
            time = elf.dayends - ( e.available_to_work % elf.minsday )
            # Adjust to toy duration elf could complete in this time
            # TODO: Could allow some "wiggle room" here to overflow slightly,
            #   as that won't cost much in productivity, but might allow better
            #   allocation of tasks.
            time = int(math.floor(time*e.productivity))
            if time < 1 or time < toys.shortest():
                # Push to next day
                e.available_to_work = ( elf.daystarts +
                    elf.minsday * (1 + e.available_to_work // elf.minsday) )
                time = elf.dayends - elf.daystarts
                time = int(math.floor(time*e.productivity))
            if time < toys.shortest():
                # Still no tasks.  Just do longest remaining
                #print("Ran out of 'small' tasks with prod =",e.productivity)
                task = toys.longest()
            else:
                # Actually only need to do a task between timeneeded and time
                timeneeded = target(toys.longest()) / e.productivity
                timeneeded = int(math.ceil( 60 * math.log(timeneeded) / math.log(1.02) ))
                timeneeded = int(math.floor(timeneeded*e.productivity))
                if time < timeneeded: # Not enough actual time, do <= time
                    task = toys.find_closest_leq(time)
                    if task == None: raise Exception("Arse!")
                else:
                    task = toys.find_closest_geq(timeneeded)
                    if task == None or task > time:
                        task = toys.find_closest_leq(time)
                        if task == None: raise Exception("Arse!")
        else:
            task = toys.longest()
        log.append( LogEntry(e.available_to_work, task, e.productivity) )
        e.do_task(e.available_to_work,task)
        toys.remove_task(task)
        
def target(size):
    return 0.25 * max(1.0, size / 11470) # 405.13

e = elf.Elf(1)
toys = TaskList(toys_list)
log = []
algo_two(e, toys, log, target)

print("Mins:",e.available_to_work, "Millions of Mins:",e.available_to_work // 1000000,
      "Years {:.2f}:".format(e.available_to_work / elf.minsday / 365))
```

# Incremental improvement?

**Update:** From the "task length against duration" plot you can see that it takes very little time to deal with these "shorter" items at the end, so this is a very incremental change...

From our first attempt, we see that we end up with the elf stuck at 0.25 productivity doing "shorter" (<= 600 mins) tasks.
   - This is a _waste_ as, once an elf is stuck at 0.25, it doesn't matter what tasks we give them, they will all happen at the same effective speed.
   - In fact, as a corner case, as the clock stops when the last toy is made, not when the elf is back to work, it makes sense to finish with a long task, all things being equal.
   - We know that the distribution of tasks forces us into the elf being stuck at low productivity: the long tasks will always wipe the elf out, and there aren't enough short tasks to fully recharge.
   - Our intuition remains that it's best to try to get the elf's productivity up at least a bit, before tackling the very long tasks.
   
We can trivially improve things by simply moving all the "shorter" tasks to being the first thing the elf does.
   - Notice that there is something interesting here.
   - To do this naively, we end up not using the whole day (as we now have no small tasks to pad things out).
   - However, this doesn't matter: doing an e.g. 400 minute task at the start, even with "padding" out to the full 600 minute day, takes a total of 600 minutes.  At the end, at 0.25 productivity, it takes 1600 minutes, so we save 1000.
   - So we could almost reverse the algorithm from before: try to keep the Elf's productivity high, and if high enough do the longest task which wouldn't push the productivity below 1.0 again, say.

```python
e = elf.Elf(1)
toys = TaskList(toys_list)
log = []

# Aggressive start:
#   - Only use tasks of length >= 274
#   - Try to keep productivity between 2 and 4
#   - For now, try to do longest task we can which keeps these invariants
#   - This has the effect of keeping the productivity rather close to 2.0...
min_length_to_use = 274
min_prod = 2.0
max_prod = 4.0

while toys.tasks_remaining() > 0:
    # If outside of sanctioned hours push to next day
    e.available_to_work = move_to_next_sanctioned_hour(e.available_to_work)
    # Time left in current day to work, adjust for productivity
    time = elf.dayends - ( e.available_to_work % elf.minsday )
    time = int(math.floor(time*e.productivity))
    if time < min_length_to_use:
        # Push to next day
        # TODO: Work on end-game here
        e.available_to_work = ( elf.daystarts +
            elf.minsday * (1 + e.available_to_work // elf.minsday) )
        time = elf.dayends - elf.daystarts
        time = int(math.floor(time*e.productivity))
    if e.productivity < min_prod:
        # Find longest task which fits in day etc...
        task = toys.find_closest_leq(time)
        if task == None or task < min_length_to_use:
            # Push to next day and try again
            # TODO: Work on end-game here
            e.available_to_work = ( elf.daystarts +
                elf.minsday * (1 + e.available_to_work // elf.minsday) )
            time = elf.dayends - elf.daystarts
            time = int(math.floor(time*e.productivity))
            task = toys.find_closest_leq(time)
            if task == None or task < min_length_to_use:
                print("Exhausted possible tasks ...")
                break
    else:
        # Productivity okay
        prodnew = e.productivity * (1.02**(time/60))
        # Solve prodnew * 0.9**badhours >= min_prod
        # <=>   badhours <= log(min_prod / prodnew) / log(0.9)
        badmins = int(math.floor( 60.0 * math.log(min_prod / prodnew) / math.log(0.9) ))
        task = toys.find_closest_leq(time + badmins)
        if task == None or task < min_length_to_use:
            # Push to next day and try again
            # TODO: Work on end-game here
            e.available_to_work = ( elf.daystarts +
                elf.minsday * (1 + e.available_to_work // elf.minsday) )
            time = elf.dayends - elf.daystarts
            time = int(math.floor(time*e.productivity))
            prodnew = e.productivity * (1.02**(time/60))
            badmins = int(math.floor( 60.0 * math.log(min_prod / prodnew) / math.log(0.9) ))
            task = toys.find_closest_leq(time + badmins)
            if task == None or task < min_length_to_use:
                print("Exhausted possible tasks 2...")
                break
    log.append( LogEntry(e.available_to_work, task, e.productivity) )
    e.do_task(e.available_to_work,task)
    toys.remove_task(task)
        
```
